# Quantum Random Number Generator in IBM Runtime

## Imports

```python
from qiskit import QuantumCircuit
from qiskit_ibm_runtime import QiskitRuntimeService, SamplerV2 as Sampler
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
```

## IBM Service

```python
service = QiskitRuntimeService()
```

## Setting Backend

```python
backend = service.least_busy(simulator=False, operational=True)
print(f"Trageting Backend: {backend.name}")
print(f"Backend Version: {backend.version}")
print(f"Number of qubits: {backend.num_qubits}")
```

## Pass Manager

```python
pm = generate_preset_pass_manager(backend=backend)
```

## Quantum Random Number Generator

```python
def generate_quantum_random_number(bits=8):

    # Create a Quantum Circuit
    qc = QuantumCircuit(bits, bits)

    # Apply Hadamard gates to all qubits
    qc.h(range(bits))

    # Measure the qubits into the classical bits
    qc.measure(range(bits), range(bits))

    sampler = Sampler(mode=backend)
    sampler.options.default_shots = 1

    isa_circuit = pm.run(qc)
    
    # Execute the circuit
    job = sampler.run([isa_circuit])
    result = job.result()

    # Extract the bitstring    
    bitstring = result[0].data.c.get_bitstrings()[0]

    # Converting binary string to a decimal int
    random_value = int(bitstring, 2)

    return random_value, bitstring

```

## Running and Results

```python
n_bits = 10
val, bits = generate_quantum_random_number(n_bits)

print(f"Generated {n_bits}-bit random number:")
print(f"Binary: {bits}")
print(f"Decimal: {val}")
```
