## Quantum Random Number Generator in AerSimulator

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

def generate_quantum_random_number(bits=8):
    """
    Generates a random integer using a quantum circuit.
    
    Args:
        bits (int): The number of bits for the random number (default 8).
        
    Returns:
        int: A random integer generated by quantum superposition.
    """
    # 1. Initialize the AerSimulator
    simulator = AerSimulator()

    # 2. Create a Quantum Circuit
    # We need 'bits' number of qubits and 'bits' number of classical bits
    qc = QuantumCircuit(bits, bits)

    # 3. Apply Hadamard gates to all qubits
    # This puts each qubit into the state |+> = (|0> + |1>) / sqrt(2)
    qc.h(range(bits))

    # 4. Measure the qubits into the classical bits
    qc.measure(range(bits), range(bits))

    # 5. Execute the circuit
    # We set shots=1 because we want one specific random outcome.
    # memory=True allows us to retrieve the exact bitstring result.
    job = simulator.run(qc, shots=1, memory=True)
    result = job.result()

    # 6. Extract the bitstring from memory
    # The result is returned as a list of strings; we take the first one.
    bitstring = result.get_memory()[0]
    
    # Convert the binary string to a decimal integer
    random_value = int(bitstring, 2)
    
    return random_value, bitstring

# Example Usage:
n_bits = 10
val, bits = generate_quantum_random_number(n_bits)

print(f"Generated {n_bits}-bit random number:")
print(f"Binary: {bits}")
print(f"Decimal: {val}")
```
